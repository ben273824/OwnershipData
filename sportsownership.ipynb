```python
names = []

Owners = data["Owner"].values.tolist() 
for i in range(len(Owners)):
    Owners[i] = Owners[i].replace(",", "")
    Owners[i] = Owners[i].replace("(", "")
    Owners[i] = Owners[i].replace(")", "")
    words = Owners[i].split()
    for word in words:
        names.append(word)

print(names)

```

```python
import pandas as pd
from bs4 import BeautifulSoup
import requests 
import time
import re

identifierMap = {"Fire":["Chicago", "Fire"], "Rapids":["Colorado", 'Rapids'], "Crew":["Columbus", "Crew"], 
                 "D.C. United":["D.C.", "United"], "Dallas":["Dallas", "FC"], "Dynamo":['Houston', "Dynamo"], "Galaxy":["Los Angeles", "Galaxy"], 
                 "Revolution":["New England", "Revolution"], "Red Bulls": ["New York", "Red Bulls"], "MetroStars":["New York", "Red Bulls"], 
                 "Salt Lake":["Salt Lake City", "FC"], "Earthquakes":["San Jose", "Earthquakes"], "Clash":["San Jose", "Earthquakes"], 
                 "Sounders":["Seattle", 'Sounders'], "Kansas City":["Kansas City", "Sporting"], "Toronto":["Toronto", "FC"]}

outcomes = pd.read_csv("/Users/benjaminklick/Documents/BigOutcomes.csv", encoding="unicode_escape")



def parseYear(string):
    tname = ["fuck","fuck"]
    for name in identifierMap.keys():
        if name in string.text:
            tname[0] = identifierMap[name][0]
            tname[1] = identifierMap[name][1]
    try:
        years = re.findall(r'(\d+)\–(\d+)', string.text)[0]
        yrs = [-1,-1]
        yrs[0] = int(years[0])
        yrs[1] = int(years[1])
        try:
            if (int(re.findall(r'(\d+)', string.nextSibling.text)[0]) == yrs[1] + 1) | (("Total" in string.nextSibling.text) | ("career" in string.nextSibling.text)):
                yrs[1] += 1
        except:
            yrs[1] += 1
    except:
        years = int(re.findall(r'(\d+)', string.text)[0])
        yrs = [years, years + 1]
    yrs = list(range(yrs[0], yrs[1]))
    return (yrs, tname)

for i in range(1996,2010):
    if (i != 2003) & (i != 2007) :
        print(i)
        url = "https://en.wikipedia.org/wiki/%s_MLS_All-Star_Game" % i
        site = requests.get(url)
        soup = BeautifulSoup(site.content)  
        soup1 = soup.find("span", id="Match_details")
        try:
            players = soup1.find_all_next('table', role="presentation")[-1]
        except:
            soup1 = soup.find("span", id="Details")
            players = soup1.find_all_next('table', role="presentation")[1]
        team = players.findChildren(["a"])
        for player in team:
            if "img" not in str(player):
                link = "https://en.wikipedia.org" + re.findall("href=\"(.+?)\"", str(player))[0]
                playersite = requests.get(link)
                playersoup = BeautifulSoup(playersite.content)
                try:
                    career = playersoup.find_all('th', text = re.compile(".*Senior [Cc]areer.*"))[0]
                except:
                    pass
                stint = career.parent.nextSibling.nextSibling
                while True:
                    try:
                        dataTeams = parseYear(stint)
                        if (i in dataTeams[0]) & (dataTeams[1][0] != "fuck"):
                            outcomes.loc[(outcomes["League"] == "MLS") & (outcomes["Year"] == i) & (outcomes["Team"] == dataTeams[1][1]) & (outcomes["City"] == dataTeams[1][0]), ["Allstars"]] = outcomes.loc[(outcomes["League"] == "MLS") & (outcomes["Year"] == i) & (outcomes["Team"] == dataTeams[1][1]) & (outcomes["City"] == dataTeams[1][0]), ["Allstars"]] + 1
                    except:
                        break
                    stint = stint.nextSibling
                    if ("Total" in stint.text) | ("career" in stint.text) | ("appearances" in stint.text):
                        break

outcomes.to_csv("/Users/benjaminklick/Documents/BigOutcomes1.csv", index=False)                
```
